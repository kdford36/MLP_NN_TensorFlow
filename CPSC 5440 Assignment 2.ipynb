```python
#import libraries
import talos
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras import backend
```

```python
#import the train and test data
import pickle

#open each file (train and test)
with open('train', 'rb') as file: 
    train_dict = pickle.load(file, encoding='bytes')
    
with open('test', 'rb') as file: 
    test_dict = pickle.load(file, encoding='bytes')

#Define X and y training sets
X_train = train_dict[b'data']
y_train = train_dict[b'coarse_labels']

#Define X and y testing sets
X_test = test_dict[b'data']
y_test = test_dict[b'coarse_labels']

#Convert labels to categorical type. We have 100 categories
y_train = tf.keras.utils.to_categorical(y_train, 100)
y_test = tf.keras.utils.to_categorical(y_test,100)
```

```python
#Define parameter dictionary
p = {
    'units': [120, 240], 
    'hidden_activations': ['relu', 'sigmoid'],
    'activation': ['softmax','sigmoid'],
    'loss': ['mse','categorical_crossentropy'],
    'optimizer':['adam', 'adagrad'],
    'batch_size':[1000,2000]
}
```

```python
#define the model
def my_model(X_train, y_train, x_val, y_val, param):
    
    #instantiate the model w/ Sequential()
    model = Sequential()
    #Add input Dense layer
    model.add(Dense(units = param['units']))
    
    #Define 5 Dense hidden layers
    model.add(Dense(units=param['units'], activation=param['hidden_activations']))
    model.add(Dense(units=param['units'], activation=param['hidden_activations']))
    model.add(Dense(units=param['units'], activation=param['hidden_activations']))
    model.add(Dense(units=param['units'], activation=param['hidden_activations']))
    model.add(Dense(units = 100, activation=param['hidden_activations']))
    #compile the model
    model.compile(loss=param['loss'],
                 optimizer=param['optimizer'],
                 metrics=['accuracy'])
    #define output of model
    out = model.fit(X_train, y_train,
                   validation_data=(x_val, y_val),
                   batch_size=param['batch_size'],
                   epochs=200,
                   verbose=1)
    #define return of model
    return out, model
```

```python
talos.Scan(X_train, y_train, p, my_model, x_val=X_test, y_val=y_test, experiment_name='talos_output')
```
